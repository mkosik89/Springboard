```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import scipy.stats

from sklearn.linear_model import LinearRegression
```

```python
n = 1000

u = np.random.uniform(0,1,size=(n,1))
u2 = np.random.uniform(0,1,size=(n,1))
noise = np.random.normal(0,3,size=(n,10))
x = np.where(u>=.5,1,-1)
x2 = np.where(u2>=.5,2,1)
y = np.random.normal(x,1)*x2*100
# y = np.random.normal(size=(n,1))
```

```python
arr = np.concatenate((u,u2,noise,y),axis=1)
# arr = np.random.normal(size=(3000,10))
```

```python
df = pd.DataFrame(arr)
```

```python
df.corr()
```

```python
lm = LinearRegression()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X = np.concatenate((np.ones((X.shape[0],1)), X), axis=1)


lm.fit(X,y)
print(lm.score(X,y))
print(lm.coef_)
print(lm.intercept_)
def objective_function(beta, X, Y):
    error = OLStest(np.matmul(X,beta), Y)
    return(error)

def OLStest(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    assert len(y_true) == len(y_pred)
    return np.sum(np.square(y_true-y_pred))

def CustomLoss(yTrue,yPred):
#     ix = (yPred*yTrue)<0
    ix = np.logical_and((yPred*yTrue)>0,np.abs(yTrue)>=np.abs(yPred))
    ix = (yPred*yTrue)>0
    n = ((yPred - yTrue)**2)*2
    y = (yPred-yTrue)**2
    out = np.where(ix,y,n)
    return np.sum(out)

def objective_function2(beta, X, Y):
    error = CustomLoss(Y, np.matmul(X,beta))
    return(error)

# You must provide a starting point at which to initialize
# the parameter search space
beta_init = np.array([1]*X.shape[1])
result = minimize(objective_function, beta_init, args=(X,y),
                  method='BFGS', options={'maxiter': 500})

# The optimal values for the input parameters are stored
# in result.x
beta_hat = result.x
print('custom beta: {}'.format(beta_hat[1:]))
print('custom intercept: {}'.format(beta_hat[0]))
print('-----------')


beta_init = np.array([0]*X.shape[1])
# beta_init = np.random.normal(size=(X.shape[1]))
result = minimize(objective_function2, beta_init, args=(X,y),
                  method='Nelder-Mead', options={'maxiter': 5000})

beta_hat = result.x
print('custom beta: {}'.format(beta_hat[1:]))
print('custom intercept: {}'.format(beta_hat[0]))
print(result.message)
print('-----------')
```
