```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib.inline
```

```python
df = pd.read_excel('gsci_data.xls',index_col=0)
df.sort_index(inplace=True)
df=df[['SPGSCITR Adj. Close']]
df.rename(columns={'SPGSCITR Adj. Close':'Close'},inplace=True)
```

```python
df.plot()
```

```python
df.loc['4/2008':'9/2008'].plot()
```

```python
df2 = df.copy()
```

```python
df2.loc['4/2008':'9/2008'][df2.loc['4/2008':'9/2008']['Close']<2000] = df2.loc['4/2008':'9/2008'][df2.loc['4/2008':'9/2008']['Close']<2000]*10
```

```python
df2.plot()
```
